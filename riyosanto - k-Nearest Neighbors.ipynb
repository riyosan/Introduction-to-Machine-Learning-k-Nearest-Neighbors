```python
##part3 standar scaling / max min scalling
for k nearest neighbours, scaling the data is very important for the training, if you recall from the previous video, knn relies of distance from 1 point to another, so if one feature coumn has a huge number, this will cause the distance to be very large
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
y = df_int['income']
X = df_int.drop(columns='income')
column_names = X.columns.values
```

```python
ss = StandardScaler()

X = ss.fit_transform(X)
```

```python
X = pd.DataFrame(X)
X.columns = column_names
X.head()
```

```python
X = pd.concat([X, df_str], axis = 1)
X.head()
```

```python
##part4 training modeling
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier().fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
##part 5 choosing an appropriate k value
```

```python
error_rate = []

for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(20,10))
plt.plot(range(1,50),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
##part6 retrain with new k value
```

# Retrain with new K value

```python
knn = KNeighborsClassifier(n_neighbors=29)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
```

```python
accuracy_score(y_test, pred)
```

```python
from sklearn.metrics import plot_confusion_matrix, classification_report
```

```python
plot_confusion_matrix(knn, X_test, y_test)
```

```python
print(classification_report(y_test,pred))
```
